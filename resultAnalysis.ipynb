```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
result_path = 'model/Final_model/eval/similarity_evaluation_evaluator1_results.csv'
result = pd.read_csv(result_path)

result.head()
```

```python
result1 = result[result['steps'] != -1]
result1.head()
```

```python
def plot_learning_curve(x, pearson, spearman, name):
    plt.figure(figsize=(6, 4))
    plt.plot(x, pearson, 'r-', label=name+'_pearson')
    plt.plot(x, spearman, 'g-', label=name+'_spearman')
    
    plt.xlabel('batch number')
    plt.ylabel('correlation')
    plt.title('Correlation between distances and true values')
    
    plt.legend()
    plt.show()
```

```python
x = np.array(range(len(result1))) * 400

plot_learning_curve(x, result1['cosine_pearson'], result1['cosine_spearman'], 'cosine')
```

```python
plot_learning_curve(x, result1['euclidean_pearson'], result1['euclidean_spearman'], 'euclidean')
```

```python
plot_learning_curve(x, result1['manhattan_pearson'], result1['manhattan_spearman'], 'manhattan')
```

```python
plot_learning_curve(x, result1['dot_pearson'], result1['dot_spearman'], 'dot')
```
